```python
Confirmed_cases=Covid_set['confirmed'].sum()
recovered_cases=Covid_set['recovered'].sum()
deaths_cases=Covid_set['deaths'].sum()
labels=['Confirmed_cases','recovered_cases','death_cases']
colors=['Red','Blue','Yellow']
```

```python
plt.pie([Confirmed_cases,recovered_cases,deaths_cases],labels=labels,colors=colors)
plt.title('Covid-19 Cases 2024')
plt.legend()
plt.show()
```
